```python
import pandas as pd
import csv

url = 'https://raw.githubusercontent.com/DataInfamous/pediatric-cancer-stalled-trials/main/data/pediatric_oncology_all_status.csv.gz'
df = pd.read_csv(url, compression='gzip')
df.columns = df.columns.str.strip()

valid_statuses = {
    'COMPLETED', 'RECRUITING', 'ACTIVE_NOT_RECRUITING', 'NOT_YET_RECRUITING',
    'TERMINATED', 'WITHDRAWN', 'SUSPENDED', 'UNKNOWN', 'NO_LONGER_AVAILABLE',
    'ENROLLING_BY_INVITATION', 'AVAILABLE', 'APPROVED_FOR_MARKETING',
    'TEMPORARILY_NOT_AVAILABLE'
}
df = df[df['Study Status'].isin(valid_statuses)].copy()

stalled_statuses = ['TERMINATED', 'WITHDRAWN', 'SUSPENDED', 'UNKNOWN', 'NO_LONGER_AVAILABLE']
df['is_stalled'] = df['Study Status'].isin(stalled_statuses)

print(f"Total trials: {len(df)}")
print(f"Stalled trials: {df['is_stalled'].sum()}")
print(f"Overall stall rate: {df['is_stalled'].mean()*100:.1f}%")
```

```python
def map_disease(cond):
    if pd.isna(cond):
        return 'unclassified'
    c = cond.lower()

    # Order matters — most specific first
    if 'ewing' in c:
        return 'Ewing'
    elif 'osteosarcoma' in c:
        return 'Osteosarcoma'
    elif 'wilms' in c:
        return 'Wilms'
    elif 'hodgkin' in c and 'non-hodgkin' not in c and 'non hodgkin' not in c:
        return 'Hodgkin'
    elif 'non-hodgkin' in c or 'non hodgkin' in c or 'nhl' in c:
        return 'NHL'
    elif 'acute myeloid' in c or 'aml' in c or 'acute myelogenous' in c:
        return 'AML'
    elif 'acute lymphoblastic' in c or 'all,' in c or c.endswith(' all') or c.startswith('all,') or 'lymphoblastic leukemia' in c:
        return 'ALL'
    elif 'neuroblastoma' in c:
        return 'Neuroblastoma'
    elif 'rhabdomyosarcoma' in c or 'soft tissue sarcoma' in c or 'sts' in c:
        return 'STS'
    elif 'glioma' in c or 'medulloblastoma' in c or 'astrocytoma' in c or 'cns ' in c or 'central nervous' in c or 'brain tumor' in c or 'brain tumour' in c:
        return 'CNS'
    else:
        return 'other'

df['disease_group'] = df['Conditions'].apply(map_disease)
print(df['disease_group'].value_counts())
print(f"\n'other' percentage: {(df['disease_group']=='other').sum()/len(df)*100:.1f}%")
```

```python
# Drop unclassified for the analysis (document in README)
analysis = df[df['disease_group'].isin(
    ['Ewing','Osteosarcoma','Wilms','Hodgkin','NHL','AML','ALL','Neuroblastoma','STS','CNS']
)]

grouped = analysis.groupby('disease_group').agg(
    total_trials=('NCT Number', 'count'),
    stalled_trials=('is_stalled', 'sum')
)
grouped['stall_rate'] = grouped['stalled_trials'] / grouped['total_trials']
grouped = grouped.sort_values('stall_rate', ascending=False)
print(grouped)
```

```python
ccss = pd.DataFrame({
    'disease': ['Ewing','Osteosarcoma','Wilms','Hodgkin','AML','NHL','Neuroblastoma','STS','ALL','CNS'],
    'incidence_rate': [3.0, 5.1, 6.2, 12.2, 8.1, 11.9, 8.5, 12.0, 33.8, 31.3],
    'non_recurrence_mortality': [30.8, 40.1, 32.0, 58.4, 35.8, 43.5, 26.4, 34.0, 28.0, 21.8]
})

merged = grouped.reset_index().merge(ccss, left_on='disease_group', right_on='disease')
print(merged[['disease','total_trials','stalled_trials','stall_rate','incidence_rate','non_recurrence_mortality']])

from scipy.stats import spearmanr

# Three-axis correlation
print("\n--- Spearman correlations ---")
rho1, p1 = spearmanr(merged['stall_rate'], merged['incidence_rate'])
print(f"Stall rate × Incidence:           rho={rho1:.3f}, p={p1:.3f}")

rho2, p2 = spearmanr(merged['stall_rate'], merged['non_recurrence_mortality'])
print(f"Stall rate × Late-effects burden: rho={rho2:.3f}, p={p2:.3f}")

rho3, p3 = spearmanr(merged['total_trials'], merged['incidence_rate'])
print(f"Total trials × Incidence:         rho={rho3:.3f}, p={p3:.3f}")

rho4, p4 = spearmanr(merged['total_trials'], merged['non_recurrence_mortality'])
print(f"Total trials × Late-effects:      rho={rho4:.3f}, p={p4:.3f}")
```

```python
def parse_age(age_str):
    if pd.isna(age_str):
        return None

    age_str = str(age_str).upper().strip()

    if age_str in ['N/A', 'NA', 'NONE', '']:
        return None

    # Map age categories to an upper age limit for pediatric focus
    if 'CHILD' in age_str and 'ADULT' in age_str and 'OLDER_ADULT' in age_str:
        return 21 # Max for combined categories if adult is present
    elif 'CHILD' in age_str and 'ADULT' in age_str:
        return 21 # Max for combined child/adult
    elif 'CHILD' in age_str:
        return 18 # Assuming child goes up to 18
    elif 'ADULT' in age_str or 'OLDER_ADULT' in age_str:
        return None # Not a pediatric trial for the purpose of df_primary

    try:
        # Original logic for specific age values (unlikely to match current data)
        if 'YEAR' in age_str:
            return float(age_str.split()[0])
        elif 'MONTH' in age_str:
            return float(age_str.split()[0]) / 12
        elif 'WEEK' in age_str:
            return float(age_str.split()[0]) / 52
        elif 'DAY' in age_str:
            return float(age_str.split()[0]) / 365
    except:
        return None

    return None
```

```python
def map_condition_expanded(cond):
    if pd.isna(cond):
        return 'other'
    cond = cond.lower()

    # Leukemia
    if 'acute lymphoblastic leukemia' in cond:
        return 'ALL'
    elif 'acute myeloid leukemia' in cond:
        return 'AML'
    elif 'leukemia' in cond:
        return 'other_leukemia'

    # Lymphoma
    elif 'hodgkin' in cond:
        return 'hodgkin'
    elif 'non-hodgkin' in cond or 'nhl' in cond:
        return 'nhl'
    elif 'lymphoma' in cond:
        return 'other_lymphoma'

    # CNS
    elif 'glioma' in cond:
        return 'glioma'
    elif 'medulloblastoma' in cond:
        return 'medulloblastoma'
    elif 'brain' in cond or 'cns' in cond:
        return 'other_cns'

    # Sarcoma
    elif 'ewing' in cond:
        return 'ewing'
    elif 'osteosarcoma' in cond:
        return 'osteosarcoma'
    elif 'sarcoma' in cond:
        return 'sts'

    # Other
    elif 'neuroblastoma' in cond:
        return 'neuroblastoma'
    elif 'wilms' in cond:
        return 'wilms'
    elif 'melanoma' in cond:
        return 'melanoma'

    return 'other'
```
